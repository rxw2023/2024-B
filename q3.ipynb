```python
import numpy as np
import matplotlib.pyplot as plt

# 参数定义
p = 200  # 成品的市场售价
m = 2  # 工序数
n = 8  # 零配件数
N = 100  # 假设的生产数量

# 零配件次品率、成本、检测成本
q_ij = [0.1] * n  # 每个零配件的次品率
c_ij = [2, 8, 12, 2, 8, 12, 8, 12]  # 零配件的成本
d_ij = [1, 1, 2, 1, 1, 2, 1, 2]  # 零配件检测成本

# 成品次品率、装配成本、检测成本
p_f = 0.1  # 成品次品率
c_f = 8  # 成品装配成本
d_f = 6  # 成品检测成本
r = 40  # 成品调换损失
d_c = 10  # 成品拆解成本

# 决策变量生成函数（遗传算法种群初始化）
def generate_population(size):
    population = []
    for _ in range(size):
        # 每个个体包括 m*n 个零配件的检测决策，1个成品检测决策，1个拆解决策
        x_ij = np.random.randint(0, 2, size=(m, n))  # 零配件检测决策
        x_f = np.random.randint(0, 2)  # 成品检测决策
        x_d = np.random.randint(0, 2)  # 拆解决策
        population.append((x_ij, x_f, x_d))
    return population

# 计算利润和销量的函数
def compute_profit_and_sales(x_ij, x_f, x_d):
    total_profit = 0
    total_sales = 0
    
    # 零配件阶段利润和次品率计算
    defect_rate = 1  # 初始次品率为 1
    for i in range(m):
        for j in range(n):
            if x_ij[i][j] == 0:
                total_profit -= c_ij[j]  # 不检测，直接产生成本
            else:
                # 检测后减少次品率
                total_profit += (1 - q_ij[j]) * (p - c_ij[j] - d_ij[j])  
                defect_rate *= (1 - q_ij[j])

    # 成品阶段利润计算
    if x_f == 0:
        total_profit += (1 - p_f) * p - c_f - r  # 不检测
        defect_rate *= p_f
    else:
        total_profit += (1 - p_f) * (p - d_f) - c_f  # 检测后利润
        defect_rate *= (1 - p_f)

    # 计算合格产品数量（销量）
    qualified_products = (1 - defect_rate) * N
    total_sales = qualified_products

    # 拆解阶段利润计算
    if x_d == 0:
        total_profit -= d_c  # 不拆解
    else:
        total_profit += sum([p_j * (p - c_ij[j]) for j, p_j in enumerate(q_ij)])  # 拆解后回收零配件

    return total_profit, total_sales
def genetic_algorithm_optimization(pop_size, generations, simulations, mutation_rate):
    # 初始化种群
    population = generate_population(pop_size)
    avg_sales_per_generation = []
    avg_profits_per_generation = []
    best_strategies = []
    best_sales_per_generation = []
    best_profits_per_generation = []

    for generation in range(generations):
        total_sales_all_strategies = []
        total_profits_all_strategies = []
        strategy_info = []
        
        # 计算每个个体的平均期望销量和平均期望利润
        for individual in population:
            x_ij, x_f, x_d = individual
            sales_list = []
            profit_list = []
            for _ in range(simulations):  # 多次模拟
                profit, sales = compute_profit_and_sales(x_ij, x_f, x_d)
                sales_list.append(sales)
                profit_list.append(profit)

            avg_sales = np.mean(sales_list)  # 计算平均期望销量
            avg_profit = np.mean(profit_list)  # 计算平均期望利润
            total_sales_all_strategies.append(avg_sales)
            total_profits_all_strategies.append(avg_profit)
            strategy_info.append((individual, avg_profit, avg_sales))

        # 找到当前代次的最佳策略
        best_individual, best_profit, best_sales = max(strategy_info, key=lambda x: x[1])
        best_strategies.append(best_individual)
        best_profits_per_generation.append(best_profit)
        best_sales_per_generation.append(best_sales)

        # 输出每代次的最优策略、最优利润和最优销量
        print(f"第 {generation + 1} 代的最优策略 (零配件检测决策 x_ij): {best_individual[0]}")
        print(f"成品检测决策 x_f: {best_individual[1]}")
        print(f"拆解决策 x_d: {best_individual[2]}")
        print(f"第 {generation + 1} 代的最优利润: {best_profit:.2f}")
        print(f"第 {generation + 1} 代的最优销量: {best_sales:.2f}")
        print('-' * 50)

        # 记录每一代次的平均期望销量和平均期望利润
        avg_sales_per_generation.append(np.mean(total_sales_all_strategies))
        avg_profits_per_generation.append(np.mean(total_profits_all_strategies))

        # 选择过程
        sorted_population = sorted(list(zip(population, total_profits_all_strategies)), key=lambda x: x[1], reverse=True)
        parent_population = [x[0] for x in sorted_population[:pop_size // 2]]

        # 交叉生成后代
        new_population = []
        for _ in range(pop_size // 2):
            parent1, parent2 = np.random.choice(len(parent_population), size=2, replace=False)
            parent1 = parent_population[parent1]
            parent2 = parent_population[parent2]

            # 交叉点的选择
            crossover_point = np.random.randint(1, n)
            offspring1_x_ij = np.vstack((parent1[0][:crossover_point], parent2[0][crossover_point:]))
            offspring2_x_ij = np.vstack((parent2[0][:crossover_point], parent1[0][crossover_point:]))

            offspring1 = (offspring1_x_ij, parent1[1], parent1[2])
            offspring2 = (offspring2_x_ij, parent2[1], parent2[2])
            new_population.extend([offspring1, offspring2])

        # 变异过程
        for i in range(len(new_population)):
            if np.random.rand() < mutation_rate:
                mutation_idx = np.random.randint(0, n)
                new_population[i][0][0][mutation_idx] = 1 - new_population[i][0][0][mutation_idx]  # 变异

        population = new_population

    return avg_sales_per_generation, avg_profits_per_generation, best_sales_per_generation, best_profits_per_generation

# 执行遗传算法
simulations = 100  # 蒙特卡洛模拟次数
avg_sales_per_generation, avg_profits_per_generation, best_sales_per_generation, best_profits_per_generation = genetic_algorithm_optimization(pop_size=10, generations=500, simulations=simulations, mutation_rate=0.1)

# 可视化平均期望利润结果
plt.plot(range(1, 501), avg_profits_per_generation)
plt.title("Average Profit per Generation")
plt.xlabel("Generation")
plt.ylabel("Average Profit")
plt.show()

# 可视化平均期望销量结果
plt.plot(range(1, 501), avg_sales_per_generation)
plt.title("Average Sales per Generation")
plt.xlabel("Generation")
plt.ylabel("Average Sales")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 参数定义
p = 200  # 成品的市场售价
m = 2  # 工序数
n = 8  # 零配件数
N = 100  # 假设的生产数量

# 零配件次品率、成本、检测成本
q_ij = [0.1] * n  # 每个零配件的次品率
c_ij = [2, 8, 12, 2, 8, 12, 8, 12]  # 零配件的成本
d_ij = [1, 1, 2, 1, 1, 2, 1, 2]  # 零配件检测成本

# 成品次品率、装配成本、检测成本
p_f = 0.1  # 成品次品率
c_f = 8  # 成品装配成本
d_f = 6  # 成品检测成本
r = 40  # 成品调换损失
d_c = 10  # 成品拆解成本

# 决策变量生成函数（遗传算法种群初始化）
def generate_population(size):
    population = []
    for _ in range(size):
        # 每个个体包括 m*n 个零配件的检测决策，1个成品检测决策，1个拆解决策
        x_ij = np.random.randint(0, 2, size=(m, n))  # 零配件检测决策
        x_f = np.random.randint(0, 2)  # 成品检测决策
        x_d = np.random.randint(0, 2)  # 拆解决策
        population.append((x_ij, x_f, x_d))
    return population

# 计算利润和销量的函数，增加了蒙特卡洛模拟
def compute_profit_and_sales(x_ij, x_f, x_d, simulations):
    total_profits = []
    total_sales = []
    
    for _ in range(simulations):  # 进行多次模拟
        total_profit = 0
        total_sales_for_simulation = 0
        defect_rate = 1  # 初始次品率为 1
        
        # 零配件阶段利润和次品率计算
        for i in range(m):
            for j in range(n):
                if x_ij[i][j] == 0:
                    total_profit -= c_ij[j]  # 不检测，直接产生成本
                else:
                    # 检测后减少次品率
                    total_profit += (1 - q_ij[j]) * (p - c_ij[j] - d_ij[j])
                    defect_rate *= (1 - q_ij[j])

        # 成品阶段利润计算
        if x_f == 0:
            total_profit += (1 - p_f) * p - c_f - r  # 不检测
            defect_rate *= p_f
        else:
            total_profit += (1 - p_f) * (p - d_f) - c_f  # 检测后利润
            defect_rate *= (1 - p_f)

        # 计算合格产品数量（销量）
        qualified_products = (1 - defect_rate) * N
        total_sales_for_simulation = qualified_products

        # 拆解阶段利润计算
        if x_d == 0:
            total_profit -= d_c  # 不拆解
        else:
            total_profit += sum([p_j * (p - c_ij[j]) for j, p_j in enumerate(q_ij)])  # 拆解后回收零配件

        total_profits.append(total_profit)
        total_sales.append(total_sales_for_simulation)

    # 返回多次模拟的平均利润和平均销量
    avg_profit = np.mean(total_profits)
    avg_sales = np.mean(total_sales)
    return avg_profit, avg_sales

# 遗传算法主流程
def genetic_algorithm_optimization(pop_size, generations, simulations, mutation_rate):
    # 初始化种群
    population = generate_population(pop_size)
    best_strategy = None
    best_profit = -np.inf
    
    for generation in range(generations):
        profits = []
        
        # 计算每个个体的平均利润和销量（蒙特卡洛模拟）
        for individual in population:
            x_ij, x_f, x_d = individual
            profit, sales = compute_profit_and_sales(x_ij, x_f, x_d, simulations)
            profits.append((individual, profit))
        
        # 找到当前种群中利润最高的个体
        current_best_individual, current_best_profit = max(profits, key=lambda x: x[1])

        # 更新最优个体
        if current_best_profit > best_profit:
            best_strategy = current_best_individual
            best_profit = current_best_profit

        # 选择过程
        sorted_population = sorted(profits, key=lambda x: x[1], reverse=True)
        parent_population = [x[0] for x in sorted_population[:pop_size // 2]]

        # 交叉生成后代
        new_population = []
        for _ in range(pop_size // 2):
            parent1, parent2 = np.random.choice(len(parent_population), size=2, replace=False)
            parent1 = parent_population[parent1]
            parent2 = parent_population[parent2]
            
            # 交叉点的选择
            crossover_point = np.random.randint(1, n)
            offspring1_x_ij = np.vstack((parent1[0][:crossover_point], parent2[0][crossover_point:]))
            offspring2_x_ij = np.vstack((parent2[0][:crossover_point], parent1[0][crossover_point:]))

            offspring1 = (offspring1_x_ij, parent1[1], parent1[2])
            offspring2 = (offspring2_x_ij, parent2[1], parent2[2])
            new_population.extend([offspring1, offspring2])

        # 变异过程
        for i in range(len(new_population)):
            if np.random.rand() < mutation_rate:
                mutation_idx = np.random.randint(0, n)
                new_population[i][0][0][mutation_idx] = 1 - new_population[i][0][0][mutation_idx]  # 变异

        population = new_population

    return best_strategy, best_profit

# 遗传代数从10到500，每隔10次输出结果
generations_range = range(10, 501, 5)
best_profits = []
best_strategies = []
simulations = 100  # 蒙特卡洛模拟的次数

for gens in generations_range:
    best_strategy, best_profit = genetic_algorithm_optimization(pop_size=10, generations=gens, simulations=simulations, mutation_rate=0.1)
    best_profits.append(best_profit)
    best_strategies.append(best_strategy)

# 可视化不同遗传代数下的最优利润变化
plt.figure(figsize=(12, 8))
plt.plot(generations_range, best_profits, color='red', linewidth=2)
plt.title("Best Profit vs Generations", fontsize=16)
plt.xlabel("Generations", fontsize=14)
plt.ylabel("Best Profit", fontsize=14)
plt.grid(True)
plt.show()

# 保存结果到Excel文件
results = []
for gens, profit, strategy in zip(generations_range, best_profits, best_strategies):
    x_ij, x_f, x_d = strategy
    results.append({
        'Generations': gens,
        'Best Profit': profit,
        'Best Strategy (x_ij)': str(x_ij),
        'Final Product Inspection (x_f)': x_f,
        'Decomposition Decision (x_d)': x_d
    })

# Convert results to a DataFrame
df = pd.DataFrame(results)

# Save to Excel using `with open` via `ExcelWriter`
with pd.ExcelWriter('genetic_algorithm_results2.xlsx') as writer:
    df.to_excel(writer, index=False)

print("Results saved in genetic_algorithm_results2.xlsx")
```
